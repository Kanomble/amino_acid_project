```python
# Parse the GenBank file
genbank_file = "../data/pGT42/pGT42_WcaJ_bla_npt_KO.gb"
record = SeqIO.read(genbank_file, "genbank")

# Initialize GenomeViz object
gv = GenomeViz()

# Create a track for the genome
track = gv.add_feature_track(record.id, segments=len(record))

# Loop over features and add to track
for feature in record.features:
    if feature.type == "CDS":
        # Get feature details
        start = int(feature.location.start)
        end = int(feature.location.end)
        strand = feature.location.strand
        gene_name = feature.qualifiers.get("gene", ["Unknown"])[0]
        
        if gene_name == "Unknown":
            gene_name = ""
        
        # Choose color based on the type
        color = "green" if strand == 1 else "red"
        
        # Add the feature to the track with color and label
        track.add_feature(start, end, strand, label=gene_name, color=color)

# Plot and save
gv.plotfig()
```

```python
from pygenomeviz import GenomeViz

# Initialize GenomeViz
genomeviz = GenomeViz(fig_width=10, feature_track_ratio=1.4)

# Example data for two gene sequences
# Gene sequence 1: Starts at 0, ends at 500
gene_seq1 = [("gene1", 0, 200, "+"), ("gene2", 250, 500, "-")]

# Gene sequence 2: Starts at 0, ends at 800
gene_seq2 = [("geneA", 0, 400, "+"), ("geneB", 500, 800, "-")]

# Add tracks for each sequence
track1 = genomeviz.add_feature_track("Sequence 1",500)
for name, start, end, strand in gene_seq1:
    track1.add_feature(start, end, 1, label=name, color="blue")

track2 = genomeviz.add_feature_track("Sequence 2",800)
for name, start, end, strand in gene_seq2:
    track2.add_feature(start, end, 1, label=name, color="green")

# Plot the visualization
#genomeviz.savefig("gene_sequences_custom_axes.png")
genomeviz.plotfig()
```
