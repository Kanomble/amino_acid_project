```python
from pygenomeviz import GenomeViz
import matplotlib.pyplot as plt
```

```python
recq = (2546165,2548012, -1)
mettrans = (2544763,2545839,1)
hy1 = (2544053,2544673,1)
moca = (2543029,2544027, 1)
meth = (2540234,2542999,1)
gzip = (2539647,2540129,1)
gzip2 = (2538672,2539487,1)
```

```python
gv = GenomeViz()
gv.set_scale_xticks(ymargin=0.20)

track = gv.add_feature_track("", 9500)
#track.add_sublabel()
lineweigth=3.0
arrow_color = "white"
special_color = "lightgreen"
label_size = 20
track.add_feature(recq[1]-recq[1], recq[1]-recq[0], recq[2], fc=arrow_color, lw=lineweigth, label="RecQ DNA Helicase", text_kws=dict(size=label_size))

end1 = recq[1]-recq[0] 
start2 = end1 + recq[0] - mettrans[1]
track.add_feature(start2, start2 + mettrans[1]-mettrans[0], mettrans[2], fc=special_color, lw=lineweigth, label="S-Methyltransferase\ndomain containing protein", text_kws=dict(size=label_size))

end2 = mettrans[0] - hy1[1]
start3 = end2 + start2 + mettrans[1]-mettrans[0]
track.add_feature(start3, start3 + hy1[1] - hy1[0], hy1[2], fc=arrow_color, lw=lineweigth, label="hypothetical protein", text_kws=dict(size=label_size))

end3 = hy1[0] - moca[1]
start4 = end3 + start3 + hy1[1] - hy1[0]
track.add_feature(start4, start4 + moca[1] - moca[0], moca[2], fc=arrow_color, lw=lineweigth, label="Gfo/Idh/MocA family protein", text_kws=dict(size=label_size))


end4 = moca[0] - meth[1]
start5 = end4 + start4 + moca[1] - moca[0]
track.add_feature(start5, start5 + meth[1] - meth[0], meth[2], fc=special_color, lw=lineweigth, label="MetH - methionine synthase", text_kws=dict(size=label_size))



end5 = meth[0] - gzip[1]
start6 = end5 + start5 + meth[1] - meth[0]
track.add_feature(start6, start6 + gzip[1] - gzip[0], gzip[2], fc=arrow_color, lw=lineweigth, label="glycine zipper\ndomain containing protein", text_kws=dict(size=label_size))


end6 = gzip[0] - gzip2[1]
start7 = end6 + start6 + gzip[1] - gzip[0]
track.add_feature(start7, start7 + gzip2[1] - gzip2[0], gzip2[2], fc=arrow_color, lw=lineweigth, label="glycine zipper\ndomain containing protein", text_kws=dict(size=label_size))



gv.plotfig()
gv.savefig("../results/figures/meth_locus.svg")
```
